```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X = mnist.data
y = mnist.target
X.shape, y.shape
```

```python
def plot(image_data):
    image = image_data.reshape(28,28)
    plt.imshow(image, cmap='binary')

plot(mnist.data[0])
plt.show()
print(mnist.target[0])
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train_5 = (y_train == '5')
y_test_5 = (y_test == '5')

'''
After y_train_5 = (y_train == '5'): You get a boolean array where:

    True = the image is a digit 5
    False = the image is NOT a digit 5
'''
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
sgd_clf.predict([X[10]])
some_digit = X[0]
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring='accuracy')
```

```python
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_train_5, y_train_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=["Not 5s", "5s"])
disp.plot()
cm
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_train_5, y_train_pred)
recall = recall_score(y_train_5, y_train_pred)
f1Score = f1_score(y_train_5, y_train_pred)
print("Precision =", precision, ", Recall =", recall, "\nF1 Score =", f1Score)
```

```python
sgd_clf.decision_function([some_digit])
```

```python
from sklearn.metrics import precision_recall_curve
#Stores the decision function based scores for all values
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method="decision_function")

precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

```python
plt.ylim(bottom=0.0, top=1.0)
plt.xlim(right=50000, left=-50000)
plt.plot(thresholds, precisions[:-1], 'r--', label='Precision', linewidth=2)
plt.plot(thresholds, recalls[:-1], 'g', linewidth=2, label = "Recall" )
plt.vlines(3000, 0,1.0, "k", "dotted", label="Threshold")
plt.show()
```

```python
plt.plot(recalls, precisions)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
fpr, tpr, thresholdss =roc_curve(y_train_5, y_scores)
plt.plot(fpr, tpr, 'r')
plt.xlim(left=0.0)
plt.ylim(top=1)
```

```python
roc_auc_score(y_train_5, y_scores)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(random_state=42)
y_probab = cross_val_predict(forest_clf, X_train, y_train_5, cv=3, method="predict_proba")
```

```python
y_scores_forest = y_probab[:, 1]
precisions_forest, recalls_forest, thresholds_forest = precision_recall_curve(y_train_5, y_scores_forest)
```
