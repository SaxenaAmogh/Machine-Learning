### This is the notebook for First Lecture on Linear Algebra for Machine Learning, part of the Math for ML series.

Below is the code showing implementation of a simple linear equation in 2 variables.

The question was:
1. MarkI Solar Panel generates 1kJ power per day.
2. MarkII Solar Panel generates 4kJ power per day. Started after a month of MarkI.
3. So when MarkII would have made same amount of power that MarkI made.

Answer is 40 days after MarkI started or 10 days after MarkII started!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 100, 10)
y1 = x
y2 = 4*(x-30)
```

```python
fig, ax = plt.subplots()
plt.title('Solar Panels Problem')
plt.xlabel('No. of Days')
plt.ylabel('Power in kJ')
ax.set_xlim([0,50])
ax.set_ylim([0,100])

ax.plot(x,y1, c='red')
ax.plot(x,y2, c='blue')

plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```
