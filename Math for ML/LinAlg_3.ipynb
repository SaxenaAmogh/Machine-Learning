#### Matrices -
- Two Dimensional array of numbers or Rank 2 Tensors.
- Denoted by uppercase, italics and bold. Elements are uppercase, italics and non-bold(scalars).
- For notation, height is given priority over weight. i.e (nrow, mcol)
- Colon represents a whole row or a column.
    - X:1 --> 1st whole col.
    - X1: --> 1st whole row

#### Generic Tensor Notation -
- Uppercase, bold, italics and Sans Serif --> X
- Number of rank is in subscript for denoting. i.e. Rank 4 will have subscript (i,j,k,l)

We take an example of Rank 4 tensor, common for images.
The 4 dimensions are - Num of Images, Height(px), Width(px), Num of Color Channels

```python
images_pt = torch.zeros([32, 28, 28, 3])
images_pt
```

```python
images_tf = tf.zeros([32, 28, 28, 3])
images_tf
```
