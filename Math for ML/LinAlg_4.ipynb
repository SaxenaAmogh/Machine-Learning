##### Basic Tensor Arithmetic
- We can do basic mathematical operations on tensors like - addition, subtraction, multiplication, division etc. 
- For these basic ops, the tensor's each element is simply operated on and changed.
- i.e. if adding 2 to a tensor, its each element will be incremented by 2. Similarly for other ops.
- In PyTorch and Tensorflow, + and * are overloaded when used. At the core - torch.add(), torch.mul(), tf.multiply() and tf.add() are being implemented.

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
X = np.array([[25,2], [5,26], [3,7]])
X_pt = torch.tensor([[25,2], [5,26], [3,7]])
X_tf = tf.Variable([[25,2], [5,26], [3,7]])
```

```python
X*2
```

```python
X*2+2
```

```python
#In PyTorch and Tensorflow, + and * are overloaded when used. At the core - torch.add(), torch.mul(), tf.multiply() and tf.add() are being implemented.
X_pt*2+2
```

```python
torch.add(torch.mul(X_pt,2 ),2)
```

```python
X_tf*2+2
```

```python
tf.add(tf.multiply(X_tf,2 ),2)
```

##### Hadamard Product
- If two tensors of the same size are being operated on with basic arithmetics, they are applied element-wise by default.
- For multiplication, tensors again are simply multiplied element-wise. This is NOT matrix multiplication, but rather called the Hadamard Product.
- The mathematical notation is <math xmlns="http://www.w3.org/1998/Math/MathML">
  <mi>A</mi>
  <mo>&#x2299;</mo>
  <mi>X</mi>
</math>

```python
X
```

```python
A = X+2
A
```

```python
A + X
```

```python
#HADAMARD
A * X
```

```python
A / X
```
