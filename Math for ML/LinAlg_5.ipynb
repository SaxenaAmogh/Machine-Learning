##### Frobenius Norm
- Analogous to L2 Norm in vectors.
- Measures the size of the matrix in terms of Euclidean Distance.
    - Its baiscally the sum of magnitude of all Vectors in the Matrix!

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
X = np.array([[1,2], [3,4]])
X
```

```python
(1**2 + 2**2 + 3**2 + 4**2)**(1/2)
```

```python
np.linalg.norm(X)
```

```python
X_pt = torch.tensor([[1,2], [3,4.]])#torch.norm() needs float type
X_pt
```

```python
torch.norm(X_pt)
```

```python
X_tf = tf.Variable([[1,2], [3,4.]])#tf.norm() also needs float type
X_tf
```

```python
tf.norm(X_tf)
```

##### Matrix Multiplication
- For 2 matrices, A and B to be multiplied, the number of columns of A must be same as the number of rows of B.
    - i.e. if A --> (mXn) and B --> (nXp) ==> A x B = C where C --> (mXp)
- *IMP* -
    - Matrix multiplication is NOT mandatorily commutative. i.e A x B may not be equal to B x A.

```python
A = np.array([[3,4], [5,6], [7,8]])
B = np.array([[1,9], [2,0]])
b = np.array([1,2]) #Vector, not a matrix
A, B, b
```

```python
np.dot(A,b)
```

```python
np.dot(A,B)
```

```python
A_pt = torch.tensor([[3,4], [5,6], [7,8]])
B_pt = torch.tensor([[1,9], [2,0]])

A_tf = tf.Variable([[3,4], [5,6], [7,8]])
B_tf = tf.Variable([[1,9], [2,0]])
b_tf = tf.Variable([1,2])
```

```python
torch.matmul(A_pt, B_pt)
```

```python
tf.matmul(A_tf, B_tf)
```

```python
tf.linalg.matvec(A_tf, b_tf)
```

##### Symmetric and Identity Matrices
- A symmetric matrix is one where :-
    - Matrix is a square matrix
    - XT = X [transpose is same as matrix]
    - Elements above the diagonal must be mirror images of those below it.

- An Identity matrix is a special case of symmetric matrices, where :-
    - Every element on the main diagonal is 1.
    - Every other element on the matrix is 0.

- It is denoted by In where n = height(or width) of matrix.
- *IMP* - Any matrix or vector multiplied by Identity matrix, results in the matrix or vector itself!

```python
X_sym = torch.tensor([[0,1,2], [1,7,8], [2,8,9]])
X_sym
```

```python
X_sym.T
```

```python
X_sym.T == X_sym
```

```python
I = np.identity(3, dtype=int)
I
```

```python
I_pt = torch.tensor([[1,0,0], [0,1,0], [0,0,1]])
I_pt
```

```python
R = torch.tensor([[25,2,5], [2,4,5]])
R
```

```python
torch.matmul(R,I_pt)
```

##### Matrix Inversion
- The inverse of a matrix M (denoted by M^-1) is a matrix that, when multiplied by M gives the *Identity Matrix*
    - i.e MM^-1 = I
- This is used to solve system of linear equations effeciently.
- *IMP* - The limitation of matrix inversion is that it does NOT work on Singular Matrices.

```python
M = np.array([[4,2], [-5,-3]])
M_pt = torch.tensor([[4,2], [-5,-3.]])
M, M_pt
```

```python
M_inv = torch.inverse(M_pt)
M_inv
```

```python
np.matmul(M_pt, M_inv)
```
