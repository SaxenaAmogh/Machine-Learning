#### Vectors -
- Next we work on "Vectors" which are One-Dimensional Tensors, or an array of numbers.
- They are arranged in an order so that we can access elements by their index.
- Vectors represent a point in space:
    - length 2 --> point in 2D Space.
    - length 3 --> location in 3D Cube.
    - length n --> location in an nD tensor.


```python
import numpy as np
import torch
import tensorflow as tf
```

```python
x = np.array([25, 2, 5])
x
```

```python
len(x)
```

```python
x.shape
#the shape is not having columns as we need to declare it in matrix style to do so. Therefore transpose also won't work on this 'x'
```

#### Vector Tranposition
- Its simply manipulating a vector by flipping itself over its diagonal.
- Reverses the shape of the vector, i.e (1,3) --> (3,1).

```python
x_t =  x.T
x_t
#As discussed, no change seen due to non-matrix declaration
```

```python
y = np.array([[25, 2, 5]])
y
```

```python
y.shape
```

```python
y_t = y.T
y_t
#Now tranpose works as expected!
```
